# Corn

```python
url = "https://waterdata.usgs.gov/nwis/dv?cb_00060=on&format=rdb&site_no=10233500&legacy=&referred_module=sw&period=&begin_date=1965-08-01&end_date=1975-09-29"
corn = pd.read_table(url,comment="#")
corn = corn.drop([0],axis=0)
corn['datetime'] = pd.to_datetime(corn['datetime'])
corn = corn.rename(columns = {'143840_00060_00003':'cfs'})
corn['cfs'] = pd.to_numeric(corn['cfs'])
corn['cfd'] = 86400*corn['cfs']
corn = corn.drop(['agency_cd','143840_00060_00003_cd'],axis=1)
corn = corn.set_index('datetime')
corn.plot()
```

```python
def f(x):
    d = {}
    d['median'] = x['cfs'].median()
    d['5%-tile'] = x['cfs'].quantile(0.05)
    d['95%-tile'] = x['cfs'].quantile(0.95)
    d['mean'] = x['cfs'].mean()
    #d['+1std'] = x['cfs'].mean() + x['cfs'].std()*1
    #d['-1std'] = x['cfs'].mean() - x['cfs'].std()*1
    return pd.Series(d, index=d.keys())


fig, ax = plt.subplots(1,1)
corndoy = corn.groupby(corn.index.dayofyear).apply(f)
corndoy.plot(ax=ax,legend=False)
ax.axvline(132,linestyle='--',alpha=0.5, color='purple',label='May 12th - Approx. Peak Flow')
ax.axvline(250,linestyle='--',alpha=0.5, color='orange',label='Sept 9th - Approx. Base Flow')
ax.scatter(corndeq.index.dayofyear, corndeq, marker=".")
ax.set_yscale('log')
ax.grid()
ax.grid(which='minor', color='grey', linestyle=':')
ax.set_ylabel('flow (cfs)')

ax.set_title(f"Corn Creek Daily Statistics {corn.first_valid_index():%Y-%m-%d} to {corn.last_valid_index():%Y-%m-%d}")
ax.set_xlabel('Day of Year')
ax.legend(loc='upper right')
```

```python
corn = corn.interpolate()
corn['water year'] = corn.index.shift(-9, freq='M').year+1 

cornwy = corn.loc['1965-10-01':].groupby('water year').sum()
cornyr = corn.loc['1966-01-01':].groupby(corn.loc['1966-01-01':].index.year).sum()
cornwy['ac-ft'] = cornwy['cfd']*2.29569e-5
cornyr['ac-ft'] = cornyr['cfd']*2.29569e-5
print(f"Avg {cornwy['ac-ft'].mean():0.0f} acft per water year")
print(f"Avg {cornyr['ac-ft'].mean():0.0f} acft per year")

#43

corn_w_ppt_wy = pd.concat([priswy.loc[43],cornwy.drop(['site_no','cfs','cfd'],axis=1)],axis=1).dropna()

import statsmodels.api as sm

x = corn_w_ppt_wy['prism_ppt_volume_af'].values
y = corn_w_ppt_wy['ac-ft'].values

X = sm.add_constant(x)
model = sm.OLS(y, X)
results = model.fit()

results.summary()
```

# Meadow

```python
url = "https://waterdata.usgs.gov/nwis/dv?cb_00060=on&format=rdb&site_no=10233000&begin_date=1920-03-01&end_date=2071-09-29"
mead = pd.read_table(url,comment="#")
mead = mead.drop([0],axis=0)
mead['datetime'] = pd.to_datetime(mead['datetime'])
mead = mead.rename(columns = {'143839_00060_00003':'cfs'})
mead['cfs'] = pd.to_numeric(mead['cfs'])
mead['cfd'] = 86400*mead['cfs']
mead = mead.drop(['agency_cd','143839_00060_00003_cd'],axis=1)
mead = mead.set_index('datetime')
mead.plot()
```

```python
def f(x):
    d = {}
    d['median'] = x['cfs'].median()
    d['5%-tile'] = x['cfs'].quantile(0.05)
    d['95%-tile'] = x['cfs'].quantile(0.95)
    d['mean'] = x['cfs'].mean()
    #d['+1std'] = x['cfs'].mean() + x['cfs'].std()*1
    #d['-1std'] = x['cfs'].mean() - x['cfs'].std()*1
    return pd.Series(d, index=d.keys())



meaddoy = mead.groupby(mead.index.dayofyear).apply(f)


fig, ax = plt.subplots(1,1)
meaddoy = mead.groupby(mead.index.dayofyear).apply(f)
meaddoy.plot(ax=ax,legend=False)
ax.axvline(132,linestyle='--',alpha=0.5, color='purple',label='May 12th - Approx. Peak Flow')
ax.axvline(250,linestyle='--',alpha=0.5, color='orange',label='Sept 9th - Approx. Base Flow')
ax.scatter(meadowdeq.index.dayofyear, meadowdeq, marker=".")
ax.set_yscale('log')
ax.grid()
ax.grid(which='minor', color='grey', linestyle=':')
ax.set_ylabel('flow (cfs)')

ax.set_title(f"Meadow Creek Daily Statistics {mead.first_valid_index():%Y-%m-%d} to {mead.last_valid_index():%Y-%m-%d}")
ax.set_xlabel('Day of Year')
ax.legend(loc='upper right')
```
